```python
excel_file_path = '04_similarity_matrix_cosine_sbert_refact.xlsx'

df = pd.read_excel(excel_file_path, index_col=0)
```

```python
with open('cosine_sbert_analysis.txt', 'w') as text_file:
    # Iterate through columns and rows and write values greater than 0.8 to the text file
    for col_label in df.columns:
        for row_label, value in df[col_label].items():
            if value > 0.8:
                text_file.write(f"{col_label} | {row_label} | {value}\n")
                print(f"{col_label} | {row_label} | {value}")
    
            
```
