```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
```

INPUT

```python
df = pd.read_csv('input/preprocessed_jobs_all.csv')
prep_titles = df['title']
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(prep_titles)
```

```python
range_n_clusters = list(range(2, 20))

silhouette_avg_scores = []
inertias = []

for num_clusters in range_n_clusters:
    kmeans = KMeans(num_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_avg_scores.append(silhouette_avg)
    inertias.append(kmeans.inertia_)
    print(f"For n_clusters = {num_clusters}, the average silhouette_score is: {silhouette_avg}")
```

```python
# Plot silhouette scores and the elbow on the same figure
plt.figure(figsize=(10, 6))

# First Y-axis for silhouette scores
ax1 = plt.gca()
ax1.plot(range_n_clusters, silhouette_avg_scores, marker='o', linestyle='-', color='#1C213F')
ax1.set_title('')
ax1.set_xlabel('Klaszterek száma')
ax1.set_ylabel('Silhouette pontszám', color='#1C213F')
ax1.tick_params(axis='y', labelcolor='#1C213F')

# Second Y-axis for elbow method
ax2 = ax1.twinx()
ax2.plot(range_n_clusters, inertias, marker='o', linestyle='-', color='#BF8F55')
ax2.set_ylabel('Tehetetlenség', color='#BF8F55')
ax2.tick_params(axis='y', labelcolor='#BF8F55')

plt.show()
```

```python
num_clusters = 10

# Perform KMeans clustering with the optimal number of clusters
kmeans = KMeans(num_clusters, random_state=42)
kmeans.fit(X)

# Add cluster labels to the DataFrame
df['cluster'] = kmeans.labels_

# Group the dataframe by cluster and print out the job titles for each cluster
grouped_df = df.groupby('cluster')
for cluster, titles in grouped_df:
    print(f"\nCluster {cluster}:")
    print(titles['title'].tolist())

# Convert the dense matrix to a NumPy array
X_array = np.asarray(X.todense())

# Fit and transform the data with t-SNE
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
X_tsne = tsne.fit_transform(X_array)

# Plot the clusters
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.6)
plt.colorbar(scatter)
plt.title('')
plt.xlabel('t-SNE jellemző 1')
plt.ylabel('t-SNE jellemző 2')
plt.show()
```
